```python
people = raw_tracks[raw_tracks['class_name'] == 'person']
crossing_tracks = pd.DataFrame()
x_thr = 510
y_thr = 400
for track_id in people['track_id'].unique():
    this_track = people[people['track_id'] == track_id]
    if (this_track['centroid_x'] < x_thr).any() and (this_track['centroid_x'] > x_thr).any():
        crossing_tracks = crossing_tracks.append(this_track)
```
